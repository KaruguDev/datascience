# Data Grouping

```python
import pandas as pd
import numpy as np
import os
```

```python
cwd = os.getcwd()

#Get File
file = cwd + '/data_source/BreadBasket_DMS.csv'

#Read CSV using Pandas
df = pd.read_csv(file)
df.head(10)
```

```python
#arrange data in ascending manner
df = df.sort_values(by='Transaction', ascending=True)
df.head()
```

```python
#arrange data in a descending manner
df = df.sort_values(by='Transaction', ascending=False)
df.head()
```

```python
#Group by qualitative attributes and count
grouped_by_item = df.groupby('Item').count()
grouped_by_item_desc = grouped_by_item.sort_values(by='Transaction', ascending=False) #descending order
grouped_by_item_desc[['Transaction']].head()
```

```python
#Group Date by group labels i.e Item
grouped_items_by_date = df['Date'].groupby(df['Item']).count()
grouped_items_by_date.head()
```

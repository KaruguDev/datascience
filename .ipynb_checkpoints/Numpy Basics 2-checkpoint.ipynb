```python
import random
```

```python
x = [1,4,5,6]
y = [23,4,12]

x = np.array(x)
y = np.array(y)

#concatenate two arrays
z = np.concatenate([x, y])

print(z)
```

```python
#autogenerated multidimensional numpy array 4*4
big_arr = []
for i in range(0,4):
    min_arr = []
    for n in range(0,4):
        num = random.randint(1,10)
        min_arr.append(num)
    big_arr.append(min_arr)

np_big_arr = np.array(big_arr)
print(np_big_arr)
```

```python
#column by row to get value
print(np_big_arr[3,0])
print(np_big_arr[1:3,1:3])
```

```python
#change a specific value
np_big_arr[3,0]=9
print(np_big_arr)
```

# Pivoting

## Deriving a new table from a existing one

* Specify index or indices to pivot on
* data is reshaped based on column values


```python
import pandas as pd
import numpy as np
import os
```

```python
#file path
cwd = os.getcwd()

#file
file = cwd + '/data_source/vgsales.csv'

#Dataframe
vgs = pd.read_csv(file)
vgs.head()
```

```python
vgsp = pd.pivot_table(vgs, index=['Rank','Name', 'Year'])
vgsp.head()
```

```python
vgsp2 = pd.pivot_table(vgs, index=['Rank','Name', 'Year'], values=['EU_Sales', 'JP_Sales'])
vgsp2.head()
```

## Apply functions

```python
vgsf = vgs[(vgs.Year >= 2008) & (vgs.Year <= 2013)]
vgsp3 = pd.pivot_table(vgsf, index=['Name'], values=['EU_Sales', 'JP_Sales'], aggfunc=np.sum)
vgsp3 = vgsp3.sort_values(by='Name', ascending=True)
vgsp3.head(1370)
```

# Matrix Arithmetic

```python
import numpy as np
```

```python
import random
```

```python
#autogenerated multidimensional numpy array 2*2
np_big_matrix = np.arange(4).reshape(2,2)
print(np_big_matrix)
```

## Multiplication

```python
#haddmard multiplication
print(np_big_matrix * np_big_matrix)
```

```python
#Element wise multiplication
print(np.multiply(np_big_matrix,np_big_matrix))
```

## Inverse

```python
from numpy.linalg import inv
```

```python
np_big_matrix_inverse = inv(np_big_matrix)
print(np_big_matrix)
```

## Transpose

```python
print(np_big_matrix.T)
```

## Solve Equations

```python
#A few rules the number of columns should be the same size
#The large matrix should be a square i.e 2*2, 4*4

a = np.random.randint(10, size=(4,4)) #4 by 4 matrix 
b = np.random.randint(10, size=4) # 4 by 1 matrix

print(a)
print(b)
```

```python
#Solve linear equations - method 1
ainv = np.linalg.inv(a)
c = ainv.dot(b)
print(c)
```

```python
#Solve linear equations - method 2
c = np.linalg.solve(a,b)
print(c)
```

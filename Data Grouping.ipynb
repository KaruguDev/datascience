# Data Grouping

```python
import pandas as pd
import numpy as np
import os
```

```python
cwd = os.getcwd()

#Get File
file = cwd + '/data_source/NBA_POTW.csv'

#Read CSV using Pandas
df = pd.read_csv(file)
df.head(10)
```

```python
#arrange data in ascending manner
df = df.sort_values(by='Seasons in league', ascending=True)
df.head()
```

```python
#arrange data in a descending manner
df = df.sort_values(by='Seasons in league', ascending=False)
df.head()
```

```python
#Group by qualitative attributes and count
grouped_by_team = df.groupby('Team').count()
grouped_by_team_desc = grouped_by_team.sort_values(by='Team', ascending=False) #descending order
grouped_by_team_desc.head()
```

```python
df.groupby(['Team', 'Player']).count()
```

```python
#Group Team by group labels i.e Player
grouped_team_by_player = df['Player'].groupby(df['Team']).count()
grouped_team_by_player.head()
```

## Lambda Functions

```python
from dateutil import parser

df['Date Formatted'] = df['Date'].apply(lambda x: parser.parse(x).strftime('%Y-%m-%d') )

df.sort_values(by='Date Formatted', ascending=True)
```

```python
group_by_team = df.groupby(['Team'])
group_by_team.size()
```

```python
group_by_teamdf = df.groupby(['Team', 'Date Formatted'])
group_by_teamdf.size()
```

```python
#counts the number of times the player's name has been found 
df['Player'].value_counts()
```

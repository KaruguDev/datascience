```python
cwd = os.getcwd()
file = cwd + '/data_source/census_data.csv'
census = pd.read_csv(file)

census.head()
```

```python
#sort columns in ascending order
c_sort = census.sort_index(axis=1, ascending=True)
c_sort.head()
```

```python
#sort by rows by specific column or colums
c_sort = census.sort_values(by="Population Density")
c_sort.head()
```

```python
c_sort = census.sort_values(by=["Area in Sq. Km.", "Population Density"])
c_sort.head()
```

## Rank

```python
#ranking the data column wise
c_rank = census.rank(axis=1)
c_rank.head()
```

```python
cr = census.iloc[:10, :-1] #rows by columns

cr['fRank'] = cr['Female'].rank(ascending=0) #rank in descending order
cr.sort_values(by='fRank')
```

# Cross Tabulation
### Cross examining data using two columns

```python
import numpy as np
import pandas as pd
import os
```

```python
cwd = os.getcwd()

#Get File
file = cwd + '/data_source/NBA_POTW.csv'

#Read CSV using Pandas
df = pd.read_csv(file)
df.head(10)
```

```python
fdf = df[(df['Draft Year'] >= 2003) & (df['Draft Year'] <= 2012)]
#fdf = df[df['Player'] == 'Michael Jordan']
cross_tab = pd.crosstab(fdf['Player'], fdf['Season short'], margins=True)
cross_tab
#cross_tab.iloc[0:,-15:]
```

```python
def perConvert(var):
    #print(var)
    return (var/float(var[-1])) * 100
```

```python
cross_tab_percentage = pd.crosstab(fdf['Player'], fdf['Season short'], margins=True).apply(perConvert, axis=1)
cross_tab_percentage
```

## Melting

### one or more columns are identifier variables while the rest are considered as measurable variables

```python
#Get File
file2 = cwd + '/data_source/fruits.csv'

#Read CSV using Pandas
bdf = pd.read_csv(file2)

#melting
print(pd.melt(bdf, id_vars=['fruit'], var_name='measurement'))
```

```python
print(pd.melt(bdf, id_vars=['fruit', 'qty'], value_name='sales'))
```
